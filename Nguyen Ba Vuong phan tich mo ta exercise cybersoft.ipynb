```python
import csv
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
np.set_printoptions(precision=3, suppress=True)
```

```python
row_list=[]

with open('nycflights.csv',mode='r',encoding='utf-8',newline='') as f:
    data_reader=csv.reader(f,delimiter=',')
    header=next(data_reader)
    print(header)
    for row in data_reader: 
        row_list.append(row)
       # print(row)    
        
len_rows = len(row_list)
print("Tổng số dòng data: ",len_rows)
 
```

```python
#chuyen khoi hanh tre
dep_delaylate=[float(row[5]) for row in row_list if row[5]!='NA']
print(dep_delaylate)
max_deplate =np.max(dep_delaylate)
print("Gia trị lớn nhất trong chuyến bay trễ :" ,max_deplate)
min_deplate =np.min(dep_delaylate)
print("Gia tri nhỏ nhất trong khởi hành trễ : " ,min_deplate)
print("Thời gian trung bình cùa khởi hành trễ là : ", np.mean(dep_delaylate))
print("Thời gian trung tâm cùa khởi hành trễ là : ", np.median(dep_delaylate))

from scipy import stats
print("Mode cho chuyến bay trễ lÀ : " ,stats.mode(dep_delaylate))
#Cách 2 tính mode 
values ,counts =np.unique(dep_delaylate, return_counts=True)
print("Gía trị mode cho chuyến bay khỡi hành trễ :",values[np.argmax(counts)])
#Phân vị
#thời gian tới nơi trễ trong khoảng 95% nhiều nhất là:
np.quantile(dep_delaylate,[0.95])
#Ví dụ chung tôi mún tặng 1 vé khứ hồi thêm cho khách hàng bị delay >100 
listkhuhoi1=np.array(dep_delaylate)
print(listkhuhoi1.size)
ds_khuhoi=100*(np.count_nonzero(listkhuhoi1>100)/listkhuhoi.size)
print(ds_khuhoi)
stats.skew(dep_delaylate)

```

```python
plt.figure(figsize=(6,6))
plt.hist(dep_delaylate,bins=-50+np.arange(30)*10,density=True)
plt.xlabel("Số liệu chuyến khởi hành trễ ")
plt.gca()
plt.show()
```
